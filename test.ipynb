```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import cv2
import os
```

```python
img = cv2.imread('face2.jpg', 0) #reading original image
img_max=np.matrix(img) #converting original image into matrix
U, S, V = np.linalg.svd(img_max) #Singular Value Decomposition for input image
print(U.shape, S.shape, V.shape)
r = [0,50, 30, 25, 20, 10] # 5 values of r for investigation
plt.figure(figsize = (16, 8)) 
for i in range(6):
  re_con = U[:, :r[i]] @ np.diag(S[:r[i]]) @ V[:r[i], :] #Reconstrunction of image
  if(i  == 0):
     plt.subplot(2, 3, i+1), plt.imshow(img, cmap = 'gray'), plt.axis('off'), plt.title("Original Image") #plot the original image
  else:
     plt.subplot(2, 3, i+1), plt.imshow(re_con, cmap = 'gray'), plt.axis('off'), plt.title("image with r =" + str(r[i])) #plot the reconstruct
```

```python
img1 = U[:, :50] @ np.diag(S[:50]) @ V[:50, :] # reconstructed image with r=50
img1_max=np.matrix(img1)


#Saving reconstructed images
matplotlib.image.imsave('r_image11.jpg', img1_max)
```
